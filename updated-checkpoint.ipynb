```python
import sqlite3
conn=sqlite3.connect("riya2.db")
c=conn.cursor()
```

```python
avg=pd.DataFrame(index = ['product','average','date'])
avg = avg.T
```

```python

year = 2007

for i in range(0,6):
    list = {}
    if i == 0:
        currentYear = year
    else :
        currentYear = year+ i +1
    dateRange = str(currentYear) + '-' + str(currentYear+1)
    a = df.loc[df['year']==(str(currentYear))]
    b = df.loc[df['year']==str(currentYear+1)]
    c=pd.concat([a,b])
    petrol = c.loc[df['petroleum_product']=='Petrol']
    turbineFuel = c.loc[df['petroleum_product']=='Aviation Turbine Fuel']
    diesel = c.loc[df['petroleum_product']=='Diesel']
    kerosene = c.loc[df['petroleum_product']=='Kerosene']
    lpg = c.loc[df['petroleum_product']=='LPG in MT']
    light= c.loc[df['petroleum_product']=='Light Diesel Oil']
    mineral = c.loc[df['petroleum_product']=='Mineral Turpentine Oil']
    petrolList = {'product':'petrol','average':petrol['sale'].mean(),'date':dateRange}
    turbineList  = {'product':'turbineFuel','average':turbineFuel['sale'].mean(),'date':dateRange}
    dieselList = {'product':'diesel','average':diesel['sale'].mean(),'date':dateRange}
    keroseneList=  {'product':'kerosene','average':kerosene['sale'].mean(),'date':dateRange}
    lpgList=  {'product':'lpg','average':lpg['sale'].mean(),'date':dateRange}
    lightList=  {'product':'light','average':light['sale'].mean(),'date':dateRange}
    mineralList=  {'product':'mineral','average':mineral['sale'].mean(),'date':dateRange}
    avg = avg.append(petrolList,ignore_index=True)
    avg = avg.append(turbineList,ignore_index=True)
    avg = avg.append(dieselList,ignore_index=True)
    avg = avg.append(keroseneList,ignore_index=True)
    avg = avg.append(lpgList,ignore_index=True)
    avg = avg.append(lightList,ignore_index=True)
    avg = avg.append(mineralList,ignore_index=True)
```
